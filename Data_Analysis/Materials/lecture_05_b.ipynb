```python
import pandas as pd
```

```python
# read dataset
data = pd.read_csv("data_lecture_05_e.csv")
print(data)
```

```python
# drop missing values
print(data.dropna())
```

```python
# replace missing values (e.g., with 0)
print(data.fillna(0))
```

```python
# fill forward
print(data.fillna(method = 'ffill'))
```

```python
# fill backward
print(data.fillna(method = 'bfill'))
```

```python
# interpolate
print(data.interpolate(method = 'linear'))
```
